```python
# Linier Regression Model Biaya Kesehatan

import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)
from sklearn.preprocessing import LabelEncoder, StandardScaler

from google.colab import drive
drive.mount('/content/drive')

folder = "/content/drive/MyDrive/[02] Work/BelajarDataScience/Dataset/"
```

```python
df=pd.read_csv("/content/drive/MyDrive/[02] Work/BelajarDataScience/Dataset/StudentPerformanceFactorsCleanNew.csv")
display(df)
```

```python
# 2. Pisahkan Feature dan Target
x = df.drop(["Exam_Score", "Academic_Status"], axis=1)
y = df["Academic_Status"]

# Encoding Kategorikal
x = pd.get_dummies(x, drop_first=True)

#Encoding Fitur Kategorikal
le = LabelEncoder()
y = le.fit_transform(y)

# 3. Split Data
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
# 5. Normalisasi (Penting untuk KNN & Logistic)

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# 6. Membuat Model
model_lr = LogisticRegression(max_iter=1000)
model_knn = KNeighborsClassifier(n_neighbors=5)
model_dt = DecisionTreeClassifier(random_state=42)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# 7. Training

model_lr.fit(x_train, y_train)
model_knn.fit(x_train, y_train)
model_dt.fit(x_train, y_train)
model_rf.fit(x_train, y_train)
```

```python
# 8. Prediksi

pred_lr = model_lr.predict(x_test)
pred_knn = model_knn.predict(x_test)
pred_dt = model_dt.predict(x_test)
pred_rf = model_rf.predict(x_test)
```

```python
# 9. Evaluasi Model

def evaluasi(nama, y_test, pred):
    acc = accuracy_score(y_test, pred)
    prec = precision_score(y_test, pred, average="weighted")
    rec = recall_score(y_test, pred, average="weighted")
    f1 = f1_score(y_test, pred, average="weighted")

    print("\n====== Hasil Evaluasi Model ======")
    print(f"Model: {nama}")
    print(f"Accuracy: {acc}")
    print(f"Precision: {prec}")
    print(f"Recall: {rec}")
    print(f"F1 Score: {f1}")

    return acc

acc_lr = evaluasi("Logistic Regression", y_test, pred_lr)
acc_knn = evaluasi("K-Nearest Neighbors", y_test, pred_knn)
acc_dt = evaluasi("Decision Tree", y_test, pred_dt)
acc_rf = evaluasi("Random Forest", y_test, pred_rf)
```

```python
# 10. Seleksi Model Terbaik

score = {
    "Logistic Regression": acc_lr,
    "K-Nearest Neighbors": acc_knn,
    "Decision Tree": acc_dt,
    "Random Forest": acc_rf
}

model_terbaik_klasifikasi = max(score, key=score.get)
print("\nModel Terbaik: ", model_terbaik_klasifikasi)
```

```python
# 11. Simpan Semua Model

joblib.dump(model_lr, folder + "model_student_logistic_regression_klasifikasi.pkl")
joblib.dump(model_knn, folder + "model_student_knn_klasifikasi.pkl")
joblib.dump(model_dt, folder + "model_student_decision_tree_klasifikasi.pkl")
joblib.dump(model_rf, folder + "model_student_random_forest_klasifikasi.pkl")
print("\nModel berhasil disimpan!")

# Simpan Scaller juga
joblib.dump(scaler, folder + "scaler.pkl")

print("\nSemua model berhasil disimpan!")
```

```python
# 12. Simpan Model Terbaik
if model_terbaik_klasifikasi == "Logistic Regression":
    model_terbaik = model_lr
elif model_terbaik_klasifikasi == "K-Nearest Neighbors":
    model_terbaik = model_knn
elif model_terbaik_klasifikasi == "Decision Tree":
    model_terbaik = model_dt
else:
    model_final_klasifikasi = model_rf

joblib.dump(model_terbaik, folder + "model_student_klasifikasi_terbaik.pkl")
print("\nModel terbaik disimpan!")
```

```python
# === Prediksi Kategori Status Kelulusan ===
df=pd.read_csv("/content/drive/MyDrive/[02] Work/BelajarDataScience/Dataset/StudentPerformanceFactorsClean.csv")
model = joblib.load(folder + "model_student_klasifikasi_terbaik.pkl")
df.info()
```

```python
# Feature referensi (harus sama dengan training)

x_ref = pd.get_dummies(
    df.drop(["Exam_Score", "Academic_Status"], axis=1), drop_first=True)

y_true = df["Academic_Status"]
```

```python
# Hitung Kesalahan Model
y_pred_all = model.predict(x_ref)

# Definisikan pemetaan dari label string ke label numerik
# Pemetaan ini selaras dengan keluaran numerik yang diharapkan dari model
label_map_reverse = {
    "Fail" : 0,
    "Remidial" : 1,
    "Pass" : 2
}

# Convert y_true (string labels) to numerical lebels
y_true_numeric = y_true.map(label_map_reverse)

accuracy = accuracy_score(y_true_numeric, y_pred_all)
error_rate = 1 - accuracy
cm = confusion_matrix(y_true_numeric, y_pred_all)

print("\n=== Evaluasi Model Global ===")
print(f"Accuracy Model: {accuracy:.3f}")
print(f"Error Rate: {error_rate:.3f}")
print("\n=== Confusion Matrix ===")
print(cm)
```
