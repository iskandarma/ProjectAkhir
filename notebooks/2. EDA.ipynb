```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/drive/MyDrive/[02] Work/BelajarDataScience/Dataset/StudentPerformanceFactors.csv")
display(df.head())
display(df.info())
```

```python
# Menghapus data null
print("DataFrame sebelum menghapus baris dengan null values:")
display(df.info())

# Menghapus baris dengan nilai null
df = df.dropna(axis=0)
print("data duplikat : ", df.duplicated().sum())

print("\nDataFrame setelah menghapus baris dengan null values:")
display(df.info())
display(df.head())
```

```python
# Menambah kolom kategori kelulusan
def kategori_kelulusan(nilai):
    if nilai >= 70:
        return "Lulus"
    elif nilai >= 60:
        return "Remidi"
    else:
        return "Tidak Lulus"

df['Kelulusan'] = df['Exam_Score'].apply(kategori_kelulusan)
display(df.head(100))
```

```python
# Distribusi data numerik (Histogram)
# untuk melihat sebaran & skewness
num_cols = [
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Physical_Activity",
    "Exam_Score"
]

df[num_cols].hist(figsize=(12,20),bins=15)

df.isnull().sum()
```

```python
#Boxplot untuk Deteksi Outlier

plt.figure(figsize=(12,8))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot Data Numerik (Outlier Detection)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribusi data kategorikal (Count Plot)

cat_cols = [
    "Parental_Involvement",
    "Access_to_Resources",
    "Extracurricular_Activities",
    "Motivation_Level",
    "Internet_Access",
    "Family_Income",
    "Teacher_Quality",
    "School_Type",
    "Peer_Influence",
    "Learning_Disabilities",
    "Parental_Education_Level",
    "Distance_from_Home",
    "Gender",
    "Kelulusan"
]

plt.figure(figsize=(17,30), constrained_layout=True)
for i, col in enumerate(cat_cols, 1):
    plt.subplot(7,2,i)
    sns.countplot(data=df, x=col)
    plt.title(f"Distribusi - {col}")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Hubungan Feature Numerik vs Target Regresi (Scatter Plot)
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.scatterplot(data=df, x="Hours_Studied", y="Exam_Score")
plt.title("Scatter Plot - Hours_Studied vs Exam_Score")

plt.subplot(2,2,2)
sns.scatterplot(data=df, x="Attendance", y="Exam_Score")
plt.title("Scatter Plot - Attendance vs Exam_Score")
plt.show()
```

```python
# 6 Perbandingan Kategorikal vs Target Numerik (Boxplot)
plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x="Hours_Studied", y="Exam_Score")
plt.title("Hours_Studied vs Exam_Score")  # Judul untuk subplot pertama
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x="Attendance", y="Exam_Score")
plt.title("Attendance vs Exam_Score")  # Judul untuk subplot pertama
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 7 Korelasi Antar Variabel Numerik (Heatmap)
plt.figure(figsize=(10,8))

sns.heatmap(
    df[num_cols].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.xticks(rotation=30)
plt.title("Heatmap Korelasi Variabel Numerik")
plt.show()
```
